##  2차원 데이터셋 

```python
from sklearn.datasets import make_blobs #2차원 데이터셋 생성
from sklearn.cluster import KMeans
```

```python
X , y = make_blobs(random_state = 2)
```

```python
X.shape
```

```python
y
```

```python
 # 군집 모델 생성 
kmeans = KMeans(n_clusters= 3)
```

```python
kmeans.fit(X)
```

```python
pred = kmeans.predict(X)
```

```python
pred
```

```python
# 중심 포인트의 좌표 
```

```python
kmeans.cluster_centers_
```

```python
# 생성한 데이터가 나눈 그룹 
# scatter 차트를 사용하는데 각각의 데이터 값들을 한번에 넣어줘야한다
# kmeans 의 단점은 모든 방향이 중요도가 같다고 생각한다
# KMeans 군집을 사용해서 나눈 그룹 시각화
center = kmeans.cluster_centers_
for n_cluster in range(3):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

```python
# KMeans
## 모든 방향의 중요도가 같다고 생각함 
## 데이터가 특정한 모양을 가지고 있는 경우 그룹을 잘 만들지 못함 
```

## 반달 데이터셋 

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200 , random_state= 0, noise= 0.05)
```

```python
for n_cluster in range(2):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters= 2)
kmeans.fit(X)
pred = kmeans.predict(X)
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

## 병합군집 

```python
# 가까이 있는 것들을 그룹으로 삼는다 
# kmeans 보다 나은 성능을 보여준다 
# 가까이 있는 것들끼리 묶어가면서 하나의 데이터로 취급해나간다. 
# 거리의 중요도를 인식하고 있는 모델은 아님 

mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=2)
pred = agg.fit_predict(X)
```

```python
pred
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

## DBSCAN 
- 밀도 기반 클러스터링
- Density Based Spatial Clustering of Applications with noise

```python
# eps = 거리 , min_samples = 최소샘플 본인 포함 , 
mglearn.plots.plot_dbscan()
```

```python
#DBSCAN은 스케일링을 진해하는 것이 좋음 
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
st_scaler.fit(X)
X_scaler = st_scaler.transform(X)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN( eps= 0.5, min_samples= 5)
pred = dbscan.fit_predict(X_scaler)
```

```python
pred
```

```python
# 데이터들이 모여있는 방향을 알 수 있음
for n_cluster in range(2):
  data = X[pred == n_cluster]
  plt.scatter(data[: , 0] , data[: , 1] , label = n_cluster)

plt.legend()
plt.show()
```
