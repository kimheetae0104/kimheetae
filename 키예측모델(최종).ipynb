```python
key_seq = []
for i in range(len(key.columns)):
    temp = pd.DataFrame(key[i])
    temp.columns=['g']
    temp[1] = range(72)
    temp = temp[[1,'g']]
    for j in range (1,len(temp)):
        seq = temp.iloc[:j+1,1].values.tolist()
        key_seq.append(seq)
    del temp

print('학습에 사용할 샘플의 개수: %d' % len(key_seq))
```

# 입력할 내용을 일정한 길이로 변경 

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = 72

key_seq = pad_sequences(key_seq, maxlen=max_len, padding='pre',dtype='float32')
```

# RNN예측

```python
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN, LSTM
from tensorflow.keras.models import Sequential
```

```python
X.shape
```

```python
model = Sequential()
model.add(SimpleRNN(64,input_shape = (71,1)))
#model.add(SimpleRNN(32))
#입력 차원 지정(셀 개수 32개), 타임스텝 길이 100, 원핫인코딩 크기 100
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='relu'))
#이진분류 이므로 1개 유닛으 가짓 Dense층 추가

#model.summary() # 전체 구조 출력
```

```python
model.compile(loss='mse',
    metrics=['mse'],)
```

```python
history = model.fit(X_train, y_train , epochs=100)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# input_height = 넘파이 배열 형식(2차원)
# n = 예측하고싶은 개월
# input_height = np.array([[50]])
def height_predict(model, input_height, n=72):
    max_len = 72
    for _ in range(len(input_height),n):# n번 반복
        encoded = pad_sequences(input_height, maxlen=max_len-1, padding='pre',dtype='float32').reshape(71,-1)
        result = model.predict([encoded.tolist()],verbose=0)
        input_height = np.hstack([input_height,result])
    return input_height
```
